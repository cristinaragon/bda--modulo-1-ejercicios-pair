## 🧠 Teoría rápida — Tuplas

Una **tupla** es una colección ordenada e **inmutable** (no se puede modificar).

```python
# Con paréntesis (forma habitual)
coordenadas = (40.4, -3.7)

# Sin paréntesis (Python también lo entiende)
coordenadas = 40.4, -3.7
```

**¿Cuándo usar tupla en vez de lista?**
- Cuando los datos no deben cambiar: coordenadas, colores RGB, fechas fijas...
- Son más rápidas y ocupan menos memoria.

**Desempaquetar** — asignar cada elemento a una variable:
```python
persona = ("Ana", 25, "Madrid")
nombre, edad, ciudad = persona   # desempaquetado
print(nombre)  # "Ana"
```

**`zip()`** — une dos listas/tuplas elemento a elemento:
```python
nombres = ["Ana", "Luis"]
edades  = [25, 30]
list(zip(nombres, edades))  # [("Ana", 25), ("Luis", 30)]
```

> 💡 `tuple(lista)` convierte una lista en tupla. `list(tupla)` hace lo contrario.

````python
# 1. Crea la tupla persona1 con estos datos: 61 kg, 1.54 m de altura, pelo marrón.
# Los tipos deben ser int, float y str respectivamente.
# Créala SIN paréntesis — Python igualmente la reconoce como tupla.
persona1 = 61, 1.54, "pelo marron"
persona1
````

````python
# 2. Crea la tupla persona2 con: 68 kg, 1.65 m, pelo rubio.
# Esta vez usa paréntesis.
persona2 = (68, 1.65, "pelo rubio")
persona2
````

````python
# 3. Crea una LISTA llamada talla con los valores: 'M', 42, 39.
talla = ["M", 42, 39]
talla
````

````python
# 4. Convierte la lista talla a tupla.
# Pista: tuple(lista) hace la conversión.
talla = tuple(talla)
talla
````

````python
# 5. Une persona1 y talla en una nueva tupla.
# Pista: las tuplas se pueden concatenar con el operador +
persona1_completa = persona1 + talla
persona1_completa
````

````python
# 6. Extrae el tercer elemento de persona2 y guárdalo en la variable color.
# Recuerda: los índices empiezan en 0, así que el tercer elemento es el índice 2.
color = persona2[2]
color
````

````python
# 7. ¿Cuántos elementos tiene persona1 ahora?
len(persona1)
````

````python
# 8. Desempaqueta persona1 asignando cada valor a su propia variable:
# peso, altura, color_pelo
# Esto se llama "desempaquetado" (unpacking) — es más limpio que acceder por índice.
peso, altura, color_pelo = persona1
print(f"Peso: {peso} kg | Altura: {altura} m | Pelo: {color_pelo}")
````

````python
# 9. Crea el zip personas combinando persona1 y persona2.
# zip() empareja los elementos por posición: (peso1, peso2), (altura1, altura2)...
# Convierte el resultado a lista para poder verlo.
personas = list(zip(persona1, persona2))
personas
````

````python
# 10. ¿Cuántos elementos tiene personas? ¿Por qué ese número y no otro?
# (Reflexiona: zip empareja por posición, así que el resultado tiene
# tantos pares como elementos tenga la tupla más corta.)
len(personas)
````

````python
# 11. Crea una tupla etiquetas con los textos: 'peso', 'altura', 'color'.
etiquetas = ("peso", "altura", "color")
etiquetas
````

````python
# 12. Crea un zip combinando etiquetas, persona1 y persona2.
# Así obtendrás tripletas: (etiqueta, valor_persona1, valor_persona2)
todo = list(zip(etiquetas, persona1, persona2))
todo
````

````python
# 13. Intenta ordenar personas con .sort()
# ¿Qué error obtienes? ¿Por qué crees que falla?
# (Pista: no se pueden comparar strings con números directamente)
try:
    personas.sort()
except TypeError as e:
    print(f"❌ Error: {e}")
    print("→ Python no puede ordenar elementos de tipos distintos (str vs float).")
````

## 📋 Resumen de lo aprendido

| Operación | Código |
|---|---|
| Crear tupla | `t = (1, 2, 3)` o `t = 1, 2, 3` |
| Acceder por índice | `t[0]`, `t[-1]` |
| Longitud | `len(t)` |
| Concatenar tuplas | `t1 + t2` |
| Desempaquetar | `a, b, c = t` |
| Lista → tupla | `tuple(lista)` |
| Tupla → lista | `list(tupla)` |
| Combinar con zip | `list(zip(t1, t2))` |

> 🔑 **Lista vs Tupla**: usa lista cuando los datos puedan cambiar; usa tupla cuando sean fijos.
