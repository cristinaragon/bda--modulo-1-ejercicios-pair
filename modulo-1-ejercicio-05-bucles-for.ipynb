## 🧠 Teoría rápida — Bucle for

Un `for` recorre cada elemento de una secuencia uno a uno:

```python
frutas = ["manzana", "pera", "kiwi"]

for fruta in frutas:
    print(fruta)   # imprime cada elemento
```

**`range()` — generar secuencias de números:**
```python
range(5)        # 0, 1, 2, 3, 4
range(2, 6)     # 2, 3, 4, 5
range(0, 10, 2) # 0, 2, 4, 6, 8  (paso de 2 en 2)
range(10, 0, -1)# 10, 9, 8... 1  (hacia atrás)
```

**List comprehension — forma compacta de crear listas:**
```python
# Forma larga:
cuadrados = []
for n in [1, 2, 3]:
    cuadrados.append(n ** 2)

# Forma compacta (list comprehension):
cuadrados = [n ** 2 for n in [1, 2, 3]]

# Con filtro:
pares = [n for n in [1, 2, 3, 4, 5] if n % 2 == 0]
```

````python
# Datos de partida — temperaturas de la semana en grados Celsius
temperaturas = [23, 12, 37, 33, 1, 6, 15]
````

````python
# 1. Clasifica las temperaturas en tres listas nuevas:
#   - bajas_temp:  temperaturas por debajo de 11 °C (incluido)
#   - medias_temp: temperaturas entre 11 y 19.9 °C
#   - altas_temp:  temperaturas de 20 °C en adelante
#
# Recorre la lista con un for y usa if/elif/else para clasificar cada valor.
# Usa .append() para añadir cada temperatura a la lista correspondiente.

bajas_temp  = []
medias_temp = []
altas_temp  = []

for temp in temperaturas:
    if temp <= 10.9:
        bajas_temp.append(temp)
    elif temp <= 19.9:
        medias_temp.append(temp)
    else:
        altas_temp.append(temp)

print("Bajas: ",  bajas_temp)
print("Medias:", medias_temp)
print("Altas: ",  altas_temp)
````

````python
# 2. Convierte las temperaturas de cada lista de Celsius a Fahrenheit.
# Fórmula: F = (C × 9/5) + 32
#
# Trabaja sobre copias para no modificar las originales.
# Usa un for anidado: primero recorre cada lista, luego cada elemento dentro de ella.
# Pista: range(len(lista)) te da los índices para modificar cada elemento in situ.

todas = bajas_temp.copy(), medias_temp.copy(), altas_temp.copy()

for lista in todas:
    for i in range(len(lista)):
        lista[i] = round((lista[i] * 9 / 5) + 32, 1)

print(todas)
````

## 🧠 Teoría rápida — range() en detalle

```python
range(inicio, fin, paso)
```

- `inicio` — primer número incluido (por defecto: 0)
- `fin`    — primer número **excluido**
- `paso`   — cuánto suma en cada vuelta (por defecto: 1)

```python
range(0, 5)      # 0 1 2 3 4
range(3, 7)      # 3 4 5 6
range(1, 11, 2)  # 1 3 5 7 9
```

````python
# 3. Imprime todos los números del rango 0 a 5 (el 6 no incluido).
for i in range(0, 6):
    print(i)
````

````python
# 4. Imprime los números del 3 al 6, ambos incluidos.
# ⚠️ range() excluye el fin — ajusta el valor en consecuencia.
for i in range(3, 7):
    print(i)
````

````python
# 5. Imprime los números entre 5 y 9, ambos excluidos.
for i in range(6, 9):
    print(i)
````

````python
# 6. Imprime un rango de 3 números que empiece en 2.
for i in range(2, 5):
    print(i)
````

````python
# 7. Imprime números empezando en 2, de 3 en 3, parando antes de llegar a 13.
for i in range(2, 13, 3):
    print(i)
````

````python
# 8. Imprime todos los números impares del 1 al 10.
# Pista: empieza en 1 y avanza de 2 en 2.
for i in range(1, 11, 2):
    print(i)
````

## 🧠 Teoría rápida — List Comprehensions

Sintaxis: `[expresion for elemento in iterable if condicion]`

```python
numeros = [1, 2, 3, 4, 5]

# Sin filtro — transforma todos:
dobles = [n * 2 for n in numeros]         # [2, 4, 6, 8, 10]

# Con filtro — solo los pares:
pares = [n for n in numeros if n % 2 == 0]  # [2, 4]

# Con condicional en la expresión (ternario):
etiquetas = ["par" if n % 2 == 0 else "impar" for n in numeros]
```

````python
# Datos de partida
lista = [2, 45, 67, 23, 45, 66, 12, 1, 34, 67, 99, 98, 67, 64, 44, 100]
````

````python
# 9. Crea una lista nueva con cada número de lista elevado al cuadrado.
# Usa list comprehension.
lista_cuadrado = [i ** 2 for i in lista]
lista_cuadrado
````

````python
# 10. Crea una lista nueva con el resto de dividir cada número entre 2.
# (El resto de un número par es 0; el de un impar es 1)
lista_resto = [i % 2 for i in lista]
lista_resto
````

````python
# 11. Crea una lista nueva con solo los elementos PARES de lista.
# Pista: añade un if al final de la comprehension.
lista_pares = [i for i in lista if i % 2 == 0]
lista_pares
````

````python
# 12. Crea una lista nueva con solo los elementos menores que 50.
lista_menores50 = [i for i in lista if i < 50]
lista_menores50
````

````python
# 13. Crea una lista donde cada elemento sea el texto 'par' o 'impar'
# según si el número correspondiente de lista es par o impar.
# Pista: usa la expresión ternaria dentro de la comprehension:
# ["par" if condicion else "impar" for ...]
lista_string = ["par" if i % 2 == 0 else "impar" for i in lista]
lista_string
````

````python
# 14. Pide al usuario un número entero positivo.
# Suma por separado los números pares e impares desde 1 hasta ese número (incluido).
# Imprime ambas sumas al final.

numero = int(input("Ingresa un número entero positivo: "))

suma_pares   = 0
suma_impares = 0

for i in range(1, numero + 1):
    if i % 2 == 0:
        suma_pares += i
    else:
        suma_impares += i

print(f"Suma de números pares:   {suma_pares}")
print(f"Suma de números impares: {suma_impares}")
````

````python
# 15. Contador de vocales.
# Pide al usuario una frase y cuenta cuántas veces aparece cada vocal (a, e, i, o, u).
# Guarda el resultado en un diccionario vocales donde la clave es la vocal
# y el valor es el número de apariciones.
#
# Pista: recorre el diccionario con .items() y usa .count() para contar.

frase  = input("Escribe una frase: ")
vocales = {"a": 0, "e": 0, "i": 0, "o": 0, "u": 0}

for vocal, _ in vocales.items():
    vocales[vocal] += frase.lower().count(vocal)

print(vocales)
````

## 📋 Resumen de lo aprendido

| Concepto | Código |
|---|---|
| Recorrer lista | `for x in lista:` |
| Recorrer por índice | `for i in range(len(lista)):` |
| Rango básico | `range(inicio, fin, paso)` |
| Añadir a lista | `lista.append(x)` |
| List comprehension | `[expr for x in lista]` |
| Con filtro | `[expr for x in lista if condicion]` |
| Con ternario | `["a" if cond else "b" for x in lista]` |
| Recorrer diccionario | `for k, v in d.items():` |
