## 🧠 Teoría rápida — Diccionarios

Un **diccionario** guarda pares `clave: valor`. La clave es única.

```python
agenda = {
    "Ana":  "612 345 678",
    "Luis": "699 111 222"
}
```

**Acceder, añadir y modificar:**
```python
agenda["Ana"]            # acceder → "612 345 678"
agenda["Marta"] = "..." # añadir nueva clave
agenda["Ana"] = "nuevo" # modificar valor existente
```

**Métodos más usados:**
```python
d.keys()                  # lista de claves
d.values()                # lista de valores
d.items()                 # lista de tuplas (clave, valor)
d.get("clave", "defecto") # accede sin error si no existe
d.update({"k": "v"})      # añade o actualiza entradas
d.pop("clave")            # elimina una clave concreta
d.popitem()               # elimina el último elemento
d.copy()                  # crea una copia independiente
d.clear()                 # vacía el diccionario
d.setdefault("k", "v")    # añade solo si la clave NO existe
```

> 💡 A diferencia de las listas, los diccionarios no usan índices numéricos. Accedes siempre por **clave**.

````python
# 1. Crea el diccionario festivos con fechas como claves y nombres como valores.
# Las claves son strings con formato 'DD/MM'.
festivos = {
    "06/01": "Epifanía del Señor",
    "15/04": "Viernes Santo",
    "15/08": "Asunción de la Virgen",
    "12/10": "Fiesta Nacional de España",
    "01/11": "Todos los Santos",
    "06/12": "Día de la Constitución Española",
    "08/12": "Inmaculada Concepción"
}
````

````python
# 2. Crea el diccionario monarcas donde la clave es el país y el valor es el nombre del monarca.
monarcas = {
    "Bélgica":     "Philippe",
    "Dinamarca":   "Margrethe II",
    "Holanda":     "Beatrix",
    "Noruega":     "Harald V",
    "España":      "Felipe VI",
    "Reino Unido": "Carlos III"
}
````

````python
# 3. ¿Qué festivo hay el 15/04?
# Accede al valor usando su clave entre corchetes.
festivos["15/04"]
````

````python
# 4. La reina Margrethe II abdicó — ahora reina Willem-Alexander en Dinamarca.
# Actualiza el valor correspondiente en el diccionario.
monarcas["Dinamarca"] = "Willem-Alexander"
monarcas
````

````python
# 5. El rey de Luxemburgo se llama Henri — no estaba en el diccionario.
# Añádelo usando .update() en lugar de la asignación directa.
# (update() es útil cuando quieres añadir o actualizar varias entradas a la vez)
monarcas.update({"Luxemburgo": "Henri"})
monarcas
````

````python
# 6. ¿Cuántos países hay definidos en monarcas?
# Pista: len() también funciona sobre diccionarios — cuenta las claves.
len(monarcas.keys())
````

````python
# 7. ¿Está 'Monaco' entre los países del diccionario?
"Monaco" in monarcas.keys()
````

````python
# 8. ¿Nos olvidamos de incluir 'Suecia'?
# Escribe la comprobación de forma que obtengas True o False como respuesta.
"Suecia" in monarcas.keys()
````

````python
# 9. Obtén una lista con todos los países del diccionario monarcas.
list(monarcas.keys())
````

````python
# 10. Obtén una lista con los nombres de todos los festivos (solo los nombres, no las fechas).
list(festivos.values())
````

````python
# 11. Imprime el diccionario festivos completo.
festivos
````

````python
# 12. Obtén una lista de TUPLAS con todos los pares (fecha, nombre_festivo).
# Pista: .items() devuelve pares clave-valor. Conviértelo a lista.
list(festivos.items())
````

````python
# 13. Copia el diccionario monarcas a monarcas2010.
# ⚠️ Usa .copy() para que sean independientes.
monarcas2010 = monarcas.copy()
monarcas2010
````

````python
# 14. Elimina España de monarcas2010 con .pop().
# .pop(clave) devuelve el valor eliminado — puedes guardarlo si lo necesitas.
monarcas2010.pop("España")
monarcas2010
````

````python
# 15. Elimina el último elemento de monarcas2010 con .popitem().
# .popitem() siempre elimina el último par añadido (orden de inserción).
monarcas2010.popitem()
monarcas2010
````

````python
# 16. Vacía completamente el diccionario monarcas2010.
monarcas2010.clear()
monarcas2010
````

````python
# 17. ¿Qué festivo hay el 06/01?
# Accede por clave directamente con corchetes.
festivos["06/01"]
````

````python
# 18. Busca qué festivo hay el 08/03.
# Esa fecha no existe en el diccionario — si accedes con corchetes dará error.
# Usa .get() con un valor por defecto: 'Día Internacional de la Mujer'.
festivos.get("08/03", "Día Internacional de la Mujer")
````

````python
# 19. Usa .setdefault() para añadir el 31/12 como 'Nochevieja' al diccionario.
# setdefault() solo añade la clave si NO existe ya — si existe, la deja igual.
# Es útil cuando quieres evitar sobreescribir datos existentes por error.
festivos.setdefault("31/12", "Nochevieja")
festivos
````

## 📋 Resumen de lo aprendido

| Operación | Código |
|---|---|
| Crear diccionario | `d = {"clave": "valor"}` |
| Acceder a un valor | `d["clave"]` |
| Acceder sin error | `d.get("clave", "defecto")` |
| Añadir / modificar | `d["nueva"] = "valor"` |
| Añadir varios | `d.update({"k1": "v1", "k2": "v2"})` |
| Añadir si no existe | `d.setdefault("k", "v")` |
| Eliminar por clave | `d.pop("clave")` |
| Eliminar el último | `d.popitem()` |
| Lista de claves | `list(d.keys())` |
| Lista de valores | `list(d.values())` |
| Lista de pares | `list(d.items())` |
| Copiar | `d.copy()` |
| Vaciar | `d.clear()` |
