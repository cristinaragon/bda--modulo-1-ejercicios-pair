## 🧠 Teoría rápida — if / elif / else

Permite que el programa elija qué hacer según una condición.

```python
temperatura = 35

if temperatura > 30:
    print("Hace calor")
elif temperatura > 20:
    print("Temperatura agradable")
else:
    print("Hace frío")
```

**Reglas importantes:**
- Solo se ejecuta el **primer bloque** cuya condición sea verdadera.
- `else` no lleva condición — es el caso por defecto si todo lo anterior falla.
- La **indentación** (4 espacios o 1 tab) es obligatoria — define qué código pertenece a cada bloque.

````python
# 1. Crea las variables de trabajo.
var1 = 8
var2 = 2
var3 = 8
````

````python
# 2. Comprueba si var1 es mayor que 6.
# Si lo es, imprime 'Es enorme'.
if var1 > 6:
    print("Es enorme")
````

````python
# 3. Comprueba si var2 es mayor que 6.
# Si lo es, imprime 'Es enorme también'.
# (Esta vez no imprimirá nada — ¿por qué?)
if var2 > 6:
    print("Es enorme también")
````

````python
# 4. Comprueba si var3 es mayor que 6.
# Si lo es, imprime dos mensajes en líneas separadas:
# 'Es enorme' y 'Pero de verdad'.
# Ambos print deben estar dentro del if (con indentación).
if var3 > 6:
    print("Es enorme")
    print("Pero de verdad")
````

````python
# 5. Comprueba si var1 es igual a var3.
# Si lo son, imprime 'Son iguales'.
if var1 == var3:
    print("Son iguales")
````

````python
# 6. Comprueba si var1 es igual a var2.
# Si lo son, imprime 'Son iguales'.
# (No imprimirá nada — ¿qué tendrías que añadir para que siempre muestre algo?)
if var1 == var2:
    print("Son iguales")
````

````python
# 7. Comprueba si var1 es mayor o igual que var2.
# Si lo es: calcula la diferencia, guárdala en dif e imprime 'La diferencia es X'.
if var1 >= var2:
    dif = var1 - var2
    print(f"La diferencia es {dif}")
````

````python
# 8. Ahora añade un else:
# Si var1 > 6 → imprime 'Es enorme'
# Si no        → imprime 'Es chiquitín'
if var1 > 6:
    print("Es enorme")
else:
    print("Es chiquitín")
````

````python
# 9. Igual para var2:
# Si var2 > 6 → 'Es enorme también'
# Si no       → 'Es chiquitín también'
if var2 > 6:
    print("Es enorme también")
else:
    print("Es chiquitín también")
````

````python
# 10. Compara var1 y var3:
# Si son iguales  → 'Son iguales'
# Si no lo son    → 'Son distintos'
if var1 == var3:
    print("Son iguales")
else:
    print("Son distintos")
````

## 🧠 Teoría rápida — elif (múltiples condiciones)

Cuando hay más de dos opciones posibles, usamos `elif` entre el `if` y el `else`:

```python
nota = 7

if nota >= 9:
    print("Sobresaliente")
elif nota >= 7:
    print("Notable")
elif nota >= 5:
    print("Aprobado")
else:
    print("Suspenso")
```

> 💡 Python evalúa las condiciones de arriba a abajo y ejecuta **solo la primera** que sea True.

````python
# 11. Programa de precios para una sala de juegos.
# Pide la edad al usuario y aplica estas tarifas:
#   - Menos de 4 años  → entrada gratis
#   - Entre 4 y 18 años → 5 euros
#   - Más de 18 años    → 10 euros
#
# ⚠️ input() siempre devuelve un string.
# Conviértelo a int con int() para poder comparar con números.

años = int(input("¿Cuántos años tienes? "))

if años < 4:
    print("Entrada gratis")
elif años <= 18:
    print("Debes pagar 5 euros")
else:
    print("Debes pagar 10 euros")
````

## 🧠 Teoría rápida — Bucle while

`while` repite un bloque de código **mientras la condición sea True**.

```python
contador = 0

while contador < 5:
    print(contador)
    contador += 1   # ⚠️ sin esto → bucle infinito
```

**Claves para no caer en un bucle infinito:**
- Define una variable de control antes del bucle.
- Asegúrate de que esa variable **cambia** dentro del bucle.
- La condición debe poder volverse `False` en algún momento.

````python
# 12. Inicia ahorros con el valor 5.
# Crea un bucle while que:
#   - Mientras ahorros sea <= 20: imprima la cantidad actual y le sume 5.
#   - Cuando supere 20, el bucle debe parar.

ahorros = 5

while ahorros <= 20:
    print(f"Los ahorros que tenemos son {ahorros}")
    ahorros += 5
````

````python
# 13. Calculadora de impuesto sobre la renta.
# Pide al usuario su ingreso anual y su estado civil (S = soltero, C = casado).
# Aplica los siguientes tramos:
#
# SOLTERO:
#   - Hasta 9.875 €     → 10%
#   - De 9.875 a 40.125 € → 9.875*10% + resto*12%
#
# CASADO:
#   - Hasta 19.850 €    → 10%
#   - De 19.850 a 80.250 € → 19.850*10% + resto*12%

ingreso_anual = float(input("Ingresa tu ingreso anual: "))
estado_civil  = input("¿Eres soltero o casado? (S/C): ")

if estado_civil.upper() == "S":
    if ingreso_anual <= 9875:
        impuesto = ingreso_anual * 0.10
    elif ingreso_anual <= 40125:
        impuesto = 9875 * 0.10 + (ingreso_anual - 9875) * 0.12
else:
    if ingreso_anual <= 19850:
        impuesto = ingreso_anual * 0.10
    elif ingreso_anual <= 80250:
        impuesto = 19850 * 0.10 + (ingreso_anual - 19850) * 0.12

print(f"Tu impuesto sobre la renta es: ${impuesto:.2f}")
````

````python
# 14. Monitor de calidad del aire.
# Pide al usuario la concentración de PM2.5 (en µg/m³).
# Clasifica la calidad según estos rangos:
#   - <= 12        → Excelente
#   - 13 a 35      → Buena
#   - 36 a 55      → Aceptable
#   - 56 a 150     → No saludable para grupos sensibles
#   - > 150        → No saludable
#
# Si la calidad es mala, muestra además un mensaje de advertencia.
# Usa try/except para evitar que el programa rompa si el usuario escribe texto.

try:
    concentracion_pm25 = float(input("Ingresa la concentración de PM2.5 en µg/m³: "))
except ValueError:
    print("❌ Error: debes ingresar un número válido.")
    concentracion_pm25 = None

if concentracion_pm25 is not None:
    if concentracion_pm25 <= 12:
        calidad_aire = "Excelente"
    elif concentracion_pm25 <= 35:
        calidad_aire = "Buena"
    elif concentracion_pm25 <= 55:
        calidad_aire = "Aceptable"
    elif concentracion_pm25 <= 150:
        calidad_aire = "No saludable para grupos sensibles"
    else:
        calidad_aire = "No saludable"

    print(f"La calidad del aire es: {calidad_aire}")

    if calidad_aire in ["No saludable para grupos sensibles", "No saludable"]:
        print("⚠️ Advertencia: la exposición prolongada puede ser perjudicial para la salud.")
````

## 📋 Resumen de lo aprendido

| Estructura | Cuándo usarla |
|---|---|
| `if condicion:` | Ejecuta el bloque si la condición es True |
| `elif condicion:` | Alternativa si la anterior no se cumplió |
| `else:` | Caso por defecto si ninguna condición fue True |
| `while condicion:` | Repite el bloque mientras la condición sea True |

> 🔑 **Indentación**: en Python el código dentro de un bloque `if` o `while` **debe** tener 4 espacios de margen. Sin ella, Python no sabe qué pertenece a qué.
