## 🧠 Teoría rápida — Listas

Una **lista** guarda varios valores en orden, en una sola variable.

```python
frutas = ["manzana", "pera", "kiwi"]  # str
numeros = [10, 20, 30]                # int
mixta = ["Ana", 25, True]             # tipos mezclados
```

**Acceder a elementos por índice** (empieza en 0):
```python
frutas[0]   # "manzana" → primero
frutas[-1]  # "kiwi"    → último
frutas[-2]  # "pera"    → penúltimo
```

**Funciones y métodos más usados:**
```python
len(lista)         # número de elementos
max(lista)         # valor máximo
min(lista)         # valor mínimo
lista.sort()       # ordena la lista (modifica el original)
lista.copy()       # crea una copia independiente
lista.clear()      # vacía la lista
lista.count(x)     # cuántas veces aparece x
lista.index(x)     # en qué posición está x
"x" in lista       # True/False → ¿está x en la lista?
```

> 💡 `list("Ana")` convierte un string en lista de caracteres: `['A', 'n', 'a']`

````python
# 1. Crea una lista llamada colores con los primeros 6 colores que te vengan a la mente.
colores = ["rojo", "amarillo", "verde", "morado", "negro", "azul"]

# 2. Crea una lista llamada buses con las líneas de autobús que pasan por tu casa.
buses = [43, 12, 14]

# 3. Crea una lista llamada temperaturas con la previsión de
# temperaturas máximas de todos los días de esta semana.
temperaturas = [18, 19, 21, 19, 11, 21, 28, 29]
````

````python
# 4. Convierte tu nombre a una lista de caracteres.
# Pista: list("texto") descompone el string letra a letra.
nombre = "Ana"
lista_nombre = list(nombre)
print(lista_nombre)
````

````python
# 5. Convierte tu apellido a una lista de caracteres.
apellido = "garcia"
lista_apellido = list(apellido)
print(lista_apellido)
````

````python
# 6. ¿Cuántos autobuses pasan por tu casa?
# Pista: len() cuenta los elementos de una lista.
print(len(buses))
````

````python
# 7. ¿Cuántos caracteres tiene tu apellido (incluyendo espacios si los hay)?
len(lista_apellido)
````

````python
# 8. ¿Cuáles son las temperaturas máxima y mínima de la semana?
# Guárdalas en temp_max y temp_min, e imprímelas con un mensaje claro.
temp_max = max(temperaturas)
temp_min = min(temperaturas)

print(f"La temperatura máxima es {temp_max} °C y la mínima es {temp_min} °C")
````

````python
# 9. ¿Hay una 'e' en tu nombre?
# Pista: usa el operador in sobre la lista de caracteres.
"e" in lista_nombre
````

````python
# 10. ¿Está el color 'rosa' en tu lista de colores?
"rosa" in colores
````

````python
# 11. ¿Tiene una 'x' tu nombre?
# Esta vez usa not in para que el resultado sea True si NO está.
"x" not in lista_nombre
````

````python
# 12. Copia el contenido de temperaturas a dos nuevas listas:
# temperaturas2 y temperaturas3.
# ⚠️ Usa .copy() — si haces temperaturas2 = temperaturas,
# ambas variables apuntarán al mismo objeto y los cambios en una afectarán a la otra.
temperaturas2 = temperaturas.copy()
temperaturas3 = temperaturas.copy()
````

````python
# 13. Vacía el contenido de temperaturas2.
# Pista: .clear() elimina todos los elementos sin borrar la variable.
temperaturas2.clear()
temperaturas2
````

````python
# 14. Imprime temperaturas y temperaturas2 para comprobar que solo
# se vació la copia, no la original.
print(temperaturas)
print(temperaturas2)
````

````python
# 15. Copia lista_nombre a lista_nombre2.
lista_nombre2 = lista_nombre.copy()
lista_nombre2
````

````python
# 16. Copia lista_apellido a lista_apellido2.
lista_apellido2 = lista_apellido.copy()
lista_apellido2
````

````python
# 17. Ordena temperaturas3 de MAYOR a MENOR.
# Pista: .sort(reverse=True) ordena en orden descendente.
# ⚠️ .sort() modifica la lista original — por eso trabajamos sobre la copia.
temperaturas3.sort(reverse=True)
temperaturas3
````

````python
# 18. Ordena alfabéticamente los caracteres de lista_nombre2 y lista_apellido2.
# Después de ejecutar, observa: ¿cómo trata Python las mayúsculas?
# (Pista: las mayúsculas van antes que las minúsculas en el orden ASCII)
lista_nombre2.sort()
lista_apellido2.sort()
print(lista_nombre2)
print(lista_apellido2)
````

````python
# 19. ¿Cuántas veces aparece la letra 'a' en tu nombre? ¿Y la 'o'?
# Pista: .count(elemento) devuelve el número de ocurrencias.
print(lista_nombre.count("a"))
print(lista_nombre.count("o"))
````

````python
# 20. Comprueba si los resultados del ejercicio 19 son los mismos en lista_nombre2.
# (Deberían serlo — mismos elementos, distinto orden.)
print(lista_nombre2.count("a"))
print(lista_nombre2.count("o"))
````

````python
# 21. ¿Cuántas veces se alcanza la temperatura máxima esta semana? ¿Y la mínima?
print(temperaturas.count(temp_max))
print(temperaturas.count(temp_min))
````

````python
# 22. Define la lista de días de la semana en inglés.
dias_en = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']
````

````python
# 23. Define la lista de temperaturas máximas de diciembre.
temperaturas_diciembre = [10, 10, 13, 14, 12, 15,
                          12, 11, 13, 17, 18, 15,
                          15, 17, 17, 13, 13, 11,
                          12, 10, 12, 14, 12, 12,
                          11, 12, 16, 16, 14, 17, 18]
````

````python
# 24. ¿Cuál es el penúltimo día de la semana?
# Pista: los índices negativos cuentan desde el final: -1 es el último, -2 el penúltimo...
dias_en[-2]
````

````python
# 25. ¿Cuál es el último elemento de la lista de temperaturas de diciembre?
temperaturas_diciembre[-1]
````

````python
# 26. ¿Está 'friday' en la lista de días? Si es así, ¿en qué posición?
# Pista: .index(elemento) devuelve el índice de la primera aparición.
print("friday" in dias_en)
print(dias_en.index("friday"))
````

````python
# 27. ¿Cuál fue la temperatura máxima en diciembre?
max(temperaturas_diciembre)
````

````python
# 28. ¿En qué posición (índice) está 'wednesday' en la lista de días?
dias_en.index("wednesday")
````

````python
# 🌟 BONUS — Imprime esta frase usando variables, no valores escritos a mano:
# "La temperatura máxima esta semana será X °C, que ocurrirá en N día(s).
#  El día más frío tendrá una temperatura de X °C, que ocurrirá en N día(s)."

print(f"La temperatura máxima esta semana será {temp_max} °C, que ocurrirá en "
      f"{temperaturas.count(temp_max)} día(s). "
      f"El día más frío tendrá una temperatura máxima de {temp_min} °C, "
      f"que ocurrirá en {temperaturas.count(temp_min)} día(s).")
````

## 📋 Resumen de lo aprendido

| Operación | Código |
|---|---|
| Crear lista | `lista = [1, 2, 3]` |

| Máximo / mínimo | `max(lista)` / `min(lista)` |
| ¿Está el elemento? | `x in lista` |
| Contar ocurrencias | `lista.count(x)` |
| Posición del elemento | `lista.index(x)` |
| Ordenar | `lista.sort()` / `lista.sort(reverse=True)` |
| Copiar | `lista.copy()` |
| Vaciar | `lista.clear()` |
| String → lista | `list("texto")` |
