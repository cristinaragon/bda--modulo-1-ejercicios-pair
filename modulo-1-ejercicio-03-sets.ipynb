## 🧠 Teoría rápida — Sets

Un **set** es una colección **sin duplicados** y **sin orden fijo**.

```python
colores = {"rojo", "azul", "verde"}    # set desde cero
unicos  = set([1, 2, 2, 3, 3, 3])      # set desde lista → {1, 2, 3}
```

**Añadir y quitar:**
```python
s.add("amarillo")           # añade un elemento
s.update(["rosa", "lila"])  # añade varios
s.remove("rojo")            # elimina (error si no existe)
s.discard("rojo")           # elimina (sin error si no existe)
```

**Operaciones entre conjuntos:**
```python
A & B                       # intersección: elementos comunes
A | B                       # unión: todos los elementos
A - B  o  A.difference(B)   # en A pero no en B
A.symmetric_difference(B)   # en A o en B, pero no en ambos
A.issubset(B)               # ¿todos los de A están en B?
A.issuperset(B)             # ¿A contiene todos los de B?
A.isdisjoint(B)             # ¿no tienen ningún elemento en común?
```

> 💡 Los sets son perfectos para eliminar duplicados de una lista: `set(lista)`

````python
# 1. Define la lista de temperaturas de diciembre (con valores repetidos).
temperaturas_diciembre = [10, 10, 13, 14, 12, 15,
                          12, 11, 13, 17, 18, 15,
                          15, 17, 17, 13, 13, 11,
                          12, 10, 12, 14, 12, 12,
                          11, 12, 16, 16, 14, 17, 18]
````

````python
# 2. Convierte la lista a un set llamado diciembre.
# Al convertir a set, los duplicados desaparecen automáticamente.
# Verás solo los valores únicos que hubo ese mes.
diciembre = set(temperaturas_diciembre)
diciembre
````

````python
# 3. Crea un set llamado colores con 6 colores de tu elección.
# Un set se crea con llaves {} — igual que un diccionario, pero sin pares clave: valor.
colores = {"rojo", "amarillo", "verde", "azul", "morado", "negro"}
````

````python
# 4. Añade 'lila' a colores.
# .add() añade un elemento. Si ya existía, no pasa nada (no se duplica).
colores.add("lila")
colores
````

````python
# 5. Añade 'azul', 'rojo' y 'granate' usando .update().
# .update() recibe un iterable (lista, set...) y añade todos sus elementos.
# Los que ya existían (azul, rojo) simplemente se ignoran.
colores.update({"azul", "rojo", "granate"})
colores
````

````python
# 6. ¿Tiene colores el valor 'verde'?
"verde" in colores
````

````python
# 7. ¿Y tiene 'ROJO' (en mayúsculas)?
# Los sets distinguen entre mayúsculas y minúsculas.
# 'rojo' y 'ROJO' son elementos distintos.
"ROJO" in colores
````

````python
# 8. Crea el set enero como copia de diciembre.
# ⚠️ Usa .copy() para que sean independientes.
enero = diciembre.copy()
enero
````

````python
# 9. Quita las temperaturas 17 y 18 de enero.
# .remove() elimina el elemento. Lanza error si no existe — usa .discard() si no estás segura.
enero.remove(17)
enero.remove(18)
enero
````

````python
# 10. ¿Cuántas temperaturas distintas hubo en diciembre?
len(diciembre)
````

````python
# 11. ¿Y cuántas en enero?
len(enero)
````

````python
# 12. ¿Qué temperaturas hay en diciembre que NO están en enero?
# .difference() devuelve los elementos que están en A pero no en B.
diciembre.difference(enero)
````

````python
# 13. ¿Qué temperaturas hay en enero que NO están en diciembre?
# (Piénsalo antes de ejecutar: ¿qué esperas obtener?)
enero.difference(diciembre)
````

````python
# 14. ¿Qué temperaturas tienen en común diciembre y enero?
# Puedes usar el operador & o el método .intersection()
diciembre & enero
````

````python
# 15. Calcula la diferencia simétrica entre colores y {'marrón', 'amarillo', 'cyan'}.
# La diferencia simétrica devuelve los elementos que están en uno u otro conjunto,
# pero NO en los dos a la vez. Es decir: los que son exclusivos de cada uno.
colores2 = {"marrón", "amarillo", "cyan"}
colores2.symmetric_difference(colores)
````

````python
# 16. ¿Son disjuntos diciembre y enero?
# Dos sets son disjuntos cuando no comparten ningún elemento.
# ¿Qué resultado esperas? ¿Por qué?
diciembre.isdisjoint(enero)
````

````python
# 17. ¿Es {'rojo', 'azul'} un superset de colores?
# Un superset contiene todos los elementos del otro conjunto (y posiblemente más).
colores3 = {"rojo", "azul"}
colores3.issuperset(colores)
````

````python
# 18. ¿Es {'rojo', 'azul'} un subset de colores?
# Un subset está completamente contenido dentro del otro conjunto.
# ¿Tiene sentido que sea subset pero no superset?
colores3.issubset(colores)
````

## 📋 Resumen de lo aprendido

| Operación | Código |
|---|---|
| Crear set | `s = {"a", "b", "c"}` |
| Lista → set (sin duplicados) | `set(lista)` |
| Añadir un elemento | `s.add(x)` |
| Añadir varios | `s.update([x, y])` |
| Eliminar (con error) | `s.remove(x)` |
| Eliminar (sin error) | `s.discard(x)` |
| Elementos comunes | `A & B` o `A.intersection(B)` |
| En A pero no en B | `A - B` o `A.difference(B)` |
| En uno pero no en ambos | `A.symmetric_difference(B)` |
| ¿A ⊂ B? | `A.issubset(B)` |
| ¿A ⊃ B? | `A.issuperset(B)` |
| ¿Sin elementos comunes? | `A.isdisjoint(B)` |
