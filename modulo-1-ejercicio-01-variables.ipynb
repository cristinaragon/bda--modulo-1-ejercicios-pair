## 🧠 Teoría rápida — Variables y tipos de datos

Una **variable** es una caja con nombre donde guardas un valor.

```python
edad = 25          # int   → número entero
precio = 9.99      # float → número con decimales
nombre = "Ana"     # str   → texto (siempre entre comillas)
activo = True      # bool  → True o False
```

**3 formas de imprimir variables con texto:**
```python
# f-string (recomendada — clara y moderna)
print(f"{nombre} tiene {edad} años")

# comas (Python añade espacios automáticamente)
print(nombre, "tiene", edad, "años")

# concatenación (hay que convertir números a texto con str())
print(nombre + " tiene " + str(edad) + " años")
```

> 💡 `type(variable)` te dice qué tipo de dato es.

````python
# 1. Crea la variable gatos con el valor entero 8.
gatos = 8

# 2. Crea la variable perros con el texto "cuatro" (6 caracteres).
perros = "cuatro"

# 3. Imprime la variable gatos.
print(gatos)

# 4. Imprime la variable perros.
print(perros)
````

````python
# 5. Imprime perros seguida de la palabra "más" y perros otra vez.
# Prueba las 3 formas: f-string, comas y concatenación.

# f-string:
print(f"{perros} más {perros}")
# comas:
print(perros, "más", perros)
# concatenación:
print(perros + " más " + perros)
````

````python
# 6. Imprime: "cuatro más cuatro es igual a 8"
# Usa las variables perros y gatos — no escribas los valores a mano.

print(f"{perros} más {perros} es igual a {gatos}")
print(perros, "más", perros, "es igual a", gatos)
print(perros + " más " + perros + " es igual a " + str(gatos))
````

````python
# 7. Imprime: "cuatro + cuatro = 8"
# Igual que antes: usa las variables, no los valores directamente.

print(f"{perros} + {perros} = {gatos}")
print(perros + " + " + perros + " = " + str(gatos))
print(perros, "+", perros, "=", gatos)
````

## 🧠 Teoría rápida — Tipos y conversión (casting)

A veces necesitas convertir un dato de un tipo a otro:


> Solución: convierte primero → `int("2") + 2`

````python
# 8. Crea la variable pizzas con valor 2 de tipo float.
pizzas = 2.0          # opción directa
# pizzas = float(2)   # opción con casting

# 9. Crea la variable magdalenas con valor 2 de tipo string.
magdalenas = "2"      # opción directa
# magdalenas = str(2) # opción con casting

# 10. Crea la variable postres con valor 0 de tipo int.
postres = 0

# 11. ¿Qué tipo de dato es la variable pizzas?
print(type(pizzas))

# 12. ¿Es magdalenas de tipo string? Compruébalo.
print(type(magdalenas))
````

## 🧠 Teoría rápida — Operaciones matemáticas

| Operador | Significado       | Ejemplo         | Resultado |
|----------|-------------------|-----------------|-----------|
| `+`      | suma              | `8 + 2`         | `10`      |
| `-`      | resta             | `8 - 2`         | `6`       |
| `*`      | multiplicación    | `8 * 2`         | `16`      |
| `/`      | división (float)  | `8 / 2`         | `4.0`     |
| `//`     | división entera   | `9 // 2`        | `4`       |
| `%`      | resto (módulo)    | `9 % 2`         | `1`       |
| `**`     | potencia          | `3 ** 2`        | `9`       |

**Operadores compuestos** (modifican la variable directamente):
```python
x += 3   # equivale a: x = x + 3
x -= 1   # equivale a: x = x - 1
x *= 2   # equivale a: x = x * 2
```

````python
# 13. Redefine postres con valor 0.
postres = 0

# 14. Calcula la suma de gatos y pizzas.
print(gatos + pizzas)

# 15. Calcula la diferencia entre gatos y pizzas.
print(gatos - pizzas)

# 16. Multiplica la cantidad de gatos por 2.
print(gatos * 2)

# 17. Crea la variable piezas: 8 veces la cantidad de pizzas.
piezas = 8 * pizzas

# 18. Aumenta piezas en 3 usando un operador compuesto.
piezas += 3
print(piezas)
````

````python
# 19. Calcula el cuadrado de 9 y guárdalo en superficie.
# Pista: elevar al cuadrado = ** 2
superficie = 9 ** 2

# 20. ¿Cuánta superficie hay por gato?
print(superficie / gatos)

# 21. ¿Cuántas piezas ENTERAS hay por gato? (sin decimales)
# Pista: usa división entera //
print(piezas // gatos)

# 22. ¿Cuántas piezas quedan si cada gato se come 21?
# Pista: usa el módulo %
print(piezas % (gatos * 21))

# 23. ¿Cuánta superficie por gato en metros cuadrados? Redondea a 1 decimal.
# Intenta calcularlo dentro del mismo print(), en una sola línea.
print(round(superficie / gatos, 1))
````

## 🧠 Teoría rápida — Comparaciones y lógica booleana

Las comparaciones devuelven `True` o `False`:

```python
5 > 3      # True
5 == 5.0   # True  (mismo valor, distinto tipo)
5 != 7     # True
```

Para combinar condiciones:
```python
# and → las DOS deben ser verdaderas
# or  → basta con que UNA sea verdadera
# not → invierte el resultado

# Forma larga:
print(x >= 2 and x < 10)

# Forma compacta (solo funciona en Python):
print(2 <= x < 10)
```

````python
# 24. ¿Es verdad que hay más gatos que pizzas?
print(gatos > pizzas)

# 25. ¿Es verdad que hay menos gatos que superficie?
print(gatos < superficie)

# 26. ¿Es verdad que 2 veces pizzas es DISTINTO de la cantidad de gatos?
print(2 * pizzas != gatos)

# 27. ¿Es FALSO que 2 veces pizzas es igual a gatos?
# Pista: usa not
print(not (2 * pizzas == gatos))

# 28. ¿piezas es mayor o igual que pizzas Y superficie es distinto de cero?
print(piezas >= pizzas and superficie != 0)

# 29. ¿piezas es mayor o igual que pizzas Y menor que superficie?
print(piezas >= pizzas and piezas < superficie)

# 30. ¿Hay más pizzas que gatos, O el residuo de piezas entre gatos es cero?
print((pizzas > gatos) or (piezas % gatos == 0))

# 31. Comprueba la condición del ejercicio 29 en una sola línea,
# sin repetir la variable piezas.
# Pista: Python permite encadenar comparaciones: a <= x < b
print(pizzas <= piezas < superficie)

# 32. Comprueba si superficie es diferente de cero SIN usar !=
# Pista: cualquier número distinto de 0 es True en Python
print(bool(superficie))
````

## 🧠 Teoría rápida — input() y métodos de strings

`input()` pausa el programa y espera que el usuario escriba algo.  
Siempre devuelve un **string**, aunque el usuario escriba un número.

```python
nombre = input("¿Cómo te llamas? ")
edad   = int(input("¿Cuántos años tienes? "))  # convertir a int si necesitas operar
```

**Métodos útiles de strings:**
```python
texto.upper()      # TODO EN MAYÚSCULAS
texto.lower()      # todo en minúsculas
texto.strip()      # elimina espacios al principio y al final
texto.replace("a", "b")  # reemplaza caracteres
```

````python
# 33. Usa input() para preguntar al usuario el día de su cumpleaños.
# Guarda la respuesta en dia_cumple.
# No conviertas a número — lo usaremos como texto.
dia_cumple = input("¿Cuál es el día de tu cumpleaños? (ej: 15): ")

# 35. Pide el mes con input() y usa .strip() para eliminar
# posibles espacios accidentales que el usuario haya escrito.
mes_cumple = input("¿En qué mes? (ej: marzo): ").strip()

# 36. Crea fecha_cumple uniendo dia_cumple y mes_cumple en un string.
fecha_cumple = dia_cumple + " de " + mes_cumple

# 37. Imprime: "Fecha de cumpleaños: " seguido de fecha_cumple.
print("Fecha de cumpleaños: " + fecha_cumple)

# 38. Imprime la variable perros en todas mayúsculas.
print(perros.upper())
````
